```python
main_df=dataset.copy()
test_df=pd.read_csv("modifiedtest.csv")
```

```python
test_df.shape
```

```python
final_df=pd.concat([main_df,test_df],axis=0)
```

```python
final_df.shape
```

```python
final_df['SalePrice']
```

```python
final_df=pd.get_dummies(final_df,columns)
```

```python
final_df.head()
```

```python
final_df =final_df.loc[:,~final_df.columns.duplicated()]
```

```python
final_df.shape
```

```python
final_df.head()
```

```python
df_train=final_df.iloc[:1460,:]
df_test=final_df.iloc[1460:,:]
```

```python
df_test['SalePrice']
```

```python
df_test.drop(['SalePrice'],axis=1,inplace=True)
```

```python
df_test.shape
```

```python
X_train=df_train.drop(['SalePrice'],axis=1)
y_train=df_train['SalePrice']
```

```python
X_train.head()
```

```python
y_train
```

```python
import xgboost as xg
regressor_xgb=xg.XGBRegressor()
```

```python
n_estimators=[100,500,900,1100]
max_depth=[2,3,5,10,15]
learning_rate=[0.05,0.1,0.15,0.20]
min_child_weight=[1,2,3,4]
base_score=[0.25,0.5,0.75,1]
hyperparameter_grid = {
    'n_estimators': n_estimators,
    'max_depth':max_depth,
    'learning_rate':learning_rate,
    'min_child_weight':min_child_weight,
    'base_score':base_score
    }
```

```python
from sklearn.model_selection import RandomizedSearchCV
random_cv=RandomizedSearchCV(estimator=regressor_xgb,param_distributions=hyperparameter_grid,cv=5,n_iter=50,scoring='neg_mean_absolute_error',
                             n_jobs=4,return_train_score = True,verbose=5,random_state=42)
```

```python
random_cv.fit(X_train,y_train)
```

```python
random_cv.best_estimator_
```

```python
regressor=xg.XGBRegressor(base_score=0.5,booster='gbtree',learning_rate=0.05,min_child_weight=3,max_depth=3,n_estimators=900)
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(df_test)
```

```python
print(y_pred)
```

```python
pred=pd.DataFrame(y_pred)
sub_df=pd.read_csv("sample_submission.csv")
datasets=pd.concat([sub_df['Id'],pred],axis=1)
datasets.columns=['Id','SalePrice']
datasets.to_csv("Final_Submission.csv",index=False)
```
